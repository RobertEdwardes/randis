```python
import pandas as pd
import geopandas as gpd
```

```python
shpFile = gpd.read_file('tl_2019_01_bg.shp')
```

```python
a = shpFile.copy()
b = shpFile.copy()
cbVE = {}
for idx, row in a.iterrows():
    polyA = row['geometry']
    geoID_A = row['GEOID']
    GEOID_array = []
    for jdx, array in b.iterrows():
        polyB = array['geometry']
        geoID_B = array['GEOID']
        if polyA.touches(polyB) is True:
            GEOID_array.append(geoID_B)
    cbVE[f'{geoID_A}'] = GEOID_array
    print(cbVE)
```
